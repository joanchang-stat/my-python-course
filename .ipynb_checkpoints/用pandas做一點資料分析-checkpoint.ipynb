```python
import pandas as pd
import matplotlib.pyplot as plt
url=("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data")
names =['sepal.length', 'sepal.width', 'petal.length', 'petal.width','class']
iris = pd.read_csv(url,header=0, sep=",",names=names)
iris.head(10)
print(iris.describe())
```

```python
iris.hist()
iris.plot(x='petal.length',y='petal.width',kind='scatter')
```

```python
iris.plot(x='sepal.length',y='sepal.width',kind='scatter')
```

```python
iris.plot(kind='kde')
```

```python
from pandas.tools.plotting import scatter_matrix
```

```python
scatter_matrix(iris,alpha= 0.2,figsize=(6,6), diagonal='kde')
```
