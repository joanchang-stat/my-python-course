## 設計自己的 Loss function = $x^3-5x^2-8x+1$
## 給定一個 learning rate =0.01

### 畫個圖看看函數長怎樣

```python
x = np.linspace(-2,5,1000000)
y = x**3-5*x**2-8*x+1
plt.plot(x,y)
```

### 微分後Loss function = $3x^2-10x-8x$

```python
x1 = np.linspace(-2,5,1000000)
y1 = 3*x**2-10*x-8
plt.plot(x1,y1)
```

```python
def f(x):
    return  x**3-5*x**2-8*x+1
def df(x):
    return  3*x**2-10*x-8
```

### x點的起始值

```python
startpoints = 1 
```

```python
learningRate = 0.01
epochs = 20 
```

```python
temptarray = np.zeros(epochs+1)
temptarray[0] = startpoints
```

### 可以設定不同 epoch 看收斂情況

```python
def GradientOptimal(startpoints,lr,epochs):
    temptarray = np.zeros(epochs+1)
    temptarray[0] = startpoints
    x = np.linspace(-2,5,1000000)
    plt.plot(x, f(x))
    for i in range(epochs):
        startpoints = startpoints - lr * df(startpoints)
        temptarray[i+1] = startpoints
        plt.scatter(startpoints,f(startpoints))
    return temptarray
```

```python
a = GradientOptimal(1,0.01,2000)
```
