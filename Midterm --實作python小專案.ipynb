<font color=Darkblue size=4 face="STCAIYUN">python的畫圖功能非常強大，因此我想藉著這次的專案，挑戰畫出特殊類型的圖，常常在做資料分析時，面對不同的資料需要呈現的圖形模式也不同，一般而言，我們將資料呈現方式分為四種類型，分別為 Comparison(例如 : Bar Chart,Column Chart,Line Chart)、Distribution(例如 :scatter chart, Histogram)、Relationship(例如 :Bubble chart)、Composition (例如 : pie chart, Area chart)，如何選擇正確的圖表呈現資料是一門很大的學問，也是資料視覺化的核心概念，因此我將介紹 Radar chart圖，3D圖，以及folium map圖，三種圖在面對不同類型的資料時，都非常實用</font>

<font color= red size=4 face="黑體"> 圖型 1 : 運用老師上課教過的學測成績資料藉由 Radar chart圖來呈現第一位學生成績</font>

```python
%matplotlib inline
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib as mpl

df1 = pd.read_csv('https://bitly.com/gradescsv')
df1['總級分'] = df1.sum(axis=1)
df1.index = range(1,101)
```

```python
df1.head(10)
```

```python
mpl.rc('font',family ='Noto Sans CJK TC')
```

```python
subject = np.array(['國文', '英文', '數學', '社會', '自然'])
boy1 = df1.loc[1,subject].values
figure = np.linspace(0, 2*np.pi, len(subject), endpoint=False)
boy1  = np.concatenate((boy1,[boy1[0]]))
figure = np.concatenate((figure,[figure[0]]))
ax1 = plt.subplot(111, polar=True)
ax1.plot(figure, boy1, linewidth=2,linestyle='solid')
ax1.fill(figure, boy1, alpha=0.65)  
ax1.set_thetagrids(figure * 180/np.pi, subject)
ax1.set_title([df1.loc[1,"姓名"]]) 
```

```python
subject = np.array(['國文', '英文', '數學', '社會', '自然'])
boy1 = df1.loc[1,subject].values
figure = np.linspace(0, 2*np.pi, len(subject), endpoint=False)
boy1  = np.concatenate((boy1,[boy1[0]]))
figure = np.concatenate((figure,[figure[0]]))
ax1 = plt.subplot(111, polar=True)
ax1.plot(figure, boy1, linewidth=2,linestyle='solid')
ax1.fill(figure, boy1, alpha=0.65)  
ax1.set_thetagrids(figure * 180/np.pi, subject)
ax1.set_title([df1.loc[1,"姓名"]]) 
ax1.set_thetamin(72)
ax1.set_thetamax(144)
plt.show()
```

<font color= blue size=3 face="黑體"> ***若只看英文和數學之間的面積，面積越大可以呈現學生英數能力越好，透過Radar chart的呈現，可以對資料有更清楚的了解

<font color= red size=4 face="黑體"> 圖型 2 用3D圖呈現學生學測成績的趨勢 </font>

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D

grades = {'X': df1.自然 , 'Y':df1.數學 , 'Z': df1.總級分 }
df=pd.DataFrame(grades)
zone = plt.figure()
ax = zone.add_subplot(111, projection='3d')
ax.scatter(df['X'], df['Y'], df['Z'], c='darkblue', s=80)
ax.set_xlabel("自然")
ax.set_ylabel("數學")
ax.set_zlabel("總級分")
ax.set_title("學測成績3D分佈圖_數學/自然/總級分")
ax.view_init(30, 160)
plt.show()
```
